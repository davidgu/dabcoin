```python
import json
import numpy as np
%matplotlib inline
#json.loads(read_data)
rows = read_data.split("\n")
rows = rows[:-1]
len(rows)
```

```python
import ast
data = []
for x in rows:
    stuff = json.loads(x)
    if stuff[2] == True:
        stuff[2] = [[1]]
    else:
        stuff[2] = [[0]]
    data.append(stuff)
```

```python
import itertools
clean_data = []
for x in data:
    merged = list(itertools.chain.from_iterable(x))
    merged = list(itertools.chain.from_iterable(merged))
    merged = [round(float(x),2) for x in merged]
    clean_data.append(np.array(merged))
data2 = clean_data
```

```python
model = Sequential()
model.add(Dense(100, init='uniform'))
model.add(Activation('relu'))
model.add(Dense(126, activation='relu',input_dim=126))
model.add(Conv1D(filters=126, kernel_size=30,activation='relu'))
model.add(Flatten()) # ADD FLATTEN LAYERSSSS

model.add(Dense(2, init='uniform'))
model.add(Activation('softmax'))

model.compile(loss='binary_crossentropy', optimizer="Adam")
model.fit(np.expand_dims(x, axis=2), keras.utils.np_utils.to_categorical(y),epochs=20)
```
